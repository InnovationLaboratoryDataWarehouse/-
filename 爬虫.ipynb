## 将评论转换为标准字符串

```python
def parseArticles(e):
    soup=BeautifulSoup(e,'lxml')
    return soup.text
```

```python
text_base=[]
for i in range(len(df3)):
    text_tem=[]
    if '全文' in df3['text'][i]:
        url='https://m.weibo.cn/statuses/extend?id='+ df3['id'][i]
        res=requests.get(url)
        jb=res.json()
        text_tem.append(jb['data']['longTextContent'])
    else:
        text_tem.append(df3['text'][i])
    text_base.append(text_tem)
```

```python
df_text=pd.DataFrame(text_base,columns=['text'])
```

```python
text_raw=list(df_text['text'].map(parseArticles))
text_raw
```
